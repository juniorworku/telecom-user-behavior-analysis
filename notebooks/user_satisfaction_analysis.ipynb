```python
# Connect to the PostgreSQL database
db = PostgresConnection(dbname='postgres_db', user='postgres_username', password='postgres_password')
db.connect()

# Execute query to retrieve data
query = "SELECT * FROM xdr_data"
result = db.execute_query(query)

# Convert result to Pandas DataFrame
df = pd.DataFrame(result, columns=[desc[0] for desc in db.cursor.description])
print(df.head())  # Display the first few rows of the DataFrame

# Close the database connection
db.close_connection()
```

```python
# Perform initial data analysis
missing_values_df = missing_values_table(df)
print("Missing Values in df:")
print(missing_values_df)
```

```python
df['HTTP DL (MB)'] = convert_bytes_to_megabytes(df, 'HTTP DL (Bytes)')

print(df.head())
```

```python
# Calculate session frequency per customer
session_frequency = df.groupby('MSISDN/Number')['Bearer Id'].count().reset_index()
session_frequency.columns = ['MSISDN/Number', 'Session Frequency']

# Calculate session duration per customer
session_duration = df.groupby('MSISDN/Number')['Dur. (ms)'].sum().reset_index()
session_duration['Session Duration (s)'] = session_duration['Dur. (ms)'] / 1000  # Convert milliseconds to seconds
session_duration.drop(columns=['Dur. (ms)'], inplace=True)

# Calculate session total traffic per customer
session_total_traffic = df.groupby('MSISDN/Number')[['Total DL (Bytes)', 'Total UL (Bytes)']].sum().reset_index()
session_total_traffic['Session Total Traffic (Bytes)'] = session_total_traffic['Total DL (Bytes)'] + session_total_traffic['Total UL (Bytes)']
session_total_traffic.drop(columns=['Total DL (Bytes)', 'Total UL (Bytes)'], inplace=True)

# Merge the metrics into a single DataFrame
engagement_features = session_frequency.merge(session_duration, on='MSISDN/Number').merge(session_total_traffic, on='MSISDN/Number')
print(engagement_features)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from scipy.spatial import distance

# Load user data
#df is user data

# Perform clustering analysis for engagement
engagement_features = df[['Session Frequency', 'Session Duration (s)', 'Session Total Traffic (Bytes)']]
engagement_kmeans = KMeans(n_clusters=3)  # Assuming 3 clusters
engagement_kmeans.fit(engagement_features)
df['Engagement Cluster'] = engagement_kmeans.labels_

# Extract centroids
less_engaged_centroid = engagement_kmeans.cluster_centers_[0]  # Centroid of less engaged cluster

# Perform clustering analysis for experience
experience_features = df[['Average TCP Retransmission', 'Average RTT', 'Average Throughput']]
experience_kmeans = KMeans(n_clusters=3)  # Assuming 3 clusters
experience_kmeans.fit(experience_features)
df['Experience Cluster'] = experience_kmeans.labels_

# Extract centroids
worst_experience_centroid = experience_kmeans.cluster_centers_[0]  # Centroid of worst experience cluster

# Compute Euclidean distances
df['Engagement Score'] = df.apply(lambda row: distance.euclidean(row[engagement_features.columns], less_engaged_centroid), axis=1)
df['Experience Score'] = df.apply(lambda row: distance.euclidean(row[experience_features.columns], worst_experience_centroid), axis=1)

# Display user data with scores
print(df[['MSISDN/Number', 'Engagement Score', 'Experience Score']])
```

```python
import numpy as np

def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

def assign_engagement_score(user_data, less_engaged_centroid):
    engagement_scores = []
    for data_point in user_data:
        distance = euclidean_distance(data_point, less_engaged_centroid)
        engagement_scores.append(distance)
    return engagement_scores

def assign_experience_score(user_data, worst_experience_centroid):
    experience_scores = []
    for data_point in user_data:
        distance = euclidean_distance(data_point, worst_experience_centroid)
        experience_scores.append(distance)
    return experience_scores

# User data is the dataset of user features
# less_engaged_centroid is the centroid of the less engaged cluster
# worst_experience_centroid is the centroid of the worst experience cluster
# Calculate engagement scores
engagement_scores = assign_engagement_score(df, less_engaged_centroid)

# Calculate experience scores
experience_scores = assign_experience_score(df, worst_experience_centroid)

# Print the engagement and experience scores
print("Engagement Scores:", engagement_scores)
print("Experience Scores:", experience_scores)
```
